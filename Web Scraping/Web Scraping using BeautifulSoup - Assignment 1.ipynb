```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
```

1) Write a python program to display all the header tags from wikipedia.org

```python
def get_url_header(url):
    res = requests.get(url)
    bs = BeautifulSoup(res.content)
    header = []
    for i in bs.find_all('span',{"class" : "mw-headline"}):
        header.append(i.text)
    print(header)
```

```python
get_url_header("https://simple.wikipedia.org/wiki/Main_Page")
```

2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year 
of release) and make data frame.

```python
def imdb_top_100(url):
    res=requests.get(url)
    sp=BeautifulSoup(res.content)
    movie=[]
    year =[]
    rating=[]
    for i in sp.find_all('h3',{"class":"lister-item-header"}):
        movie.append(i.text.replace("\n","").strip().split('(')[0].split('.')[1])
    for j in sp.find_all('span',{"class":"lister-item-year text-muted unbold"}):
        year.append(j.text.replace("\n","").strip("()"))
    for k in sp.find_all('div',{"class":"ipl-rating-star small"}):
        rating.append(k.text.replace("\n",""))

    df=pd.DataFrame({"Movie" :movie,"Year of Release":year,"Rating":rating})
    return df
```

```python
imdb_top_100("https://www.imdb.com/list/ls091520106/")
```

3) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL.

```python
def dineout_restaurant_names(url):
    res=requests.get(url)
    sp = BeautifulSoup(res.content)
    name = []
    rating = []
    location = []
    img_url =[]
    cuisine =[]
    for i in sp.find_all('div',{"class":"restnt-info cursor"}):
        name.append(i.text.strip().split(',')[0])
    for j in sp.find_all('div',{"class":["restnt-rating rating-4","restnt-rating rating-5"]}):
        rating.append(j.text)
    for k in sp.find_all('div',{"class":"restnt-loc ellipsis"}):
        location.append(k.text)
    for l in sp.find_all('img',class_="no-img"):
        img_url.append(l["data-src"])
    for m in sp.find_all('span',{"class":"double-line-ellipsis"}):
        cuisine.append(m.text.strip().split("|")[1])
    df1=pd.DataFrame(data={"Restaurant name":name,"Cuisine":cuisine,"Location":location,"Ratings":rating,"Image URL":img_url})
    return df1
```

```python
dineout_restaurant_names("https://www.dineout.co.in/delhi-restaurants")
```

4) Write s python program to display list of respected former finance minister of India(i.e. 
Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make 
data frame.

```python
def president_of_india(url):
    response = requests.get(url)
    sp = BeautifulSoup(response.content)
    president_name = []
    president_tenure =[]
    for i in sp.find_all("div",class_ ='president-listing'):
        president_name.append(i.text.strip().split("\n")[0])
    for j in sp.find_all("div",class_='president-listing'):
        president_tenure.append(j.text.strip().split("\n")[1])
    df2=pd.DataFrame({"Name of President" :president_name,"Tenure of President":president_tenure})
    return df2 
```

```python
president_of_india("https://presidentofindia.nic.in/former-presidents")
```

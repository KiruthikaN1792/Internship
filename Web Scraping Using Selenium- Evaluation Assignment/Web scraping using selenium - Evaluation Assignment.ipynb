Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
salary filter. 

You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results. 

You have to scrape the job-title, job-location, company name, experience required. 

The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs 

The task will be done as shown in the below steps: 

1. first get the web page https://www.naukri.com/
2. Enter ‚ÄúData Scientist‚Äù in ‚ÄúSkill, Designations, and Companies‚Äù field. 
3. Then click the search button. 
4. Then apply the location filter and salary filter by checking the respective boxes 
5. Then scrape the data for the first 10 jobs results you get. 
6. Finally create a dataframe of the scraped data.

```python
#Instantiate the webdriver for Chrome
driver=webdriver.Chrome()
```

```python
#Get the URL using get() method
driver.get("https://www.naukri.com/")
```

```python
#Maximize the window size of chrome
driver.maximize_window()
```

```python
#Enter the job title 'Data Scientist' in Job-title field
designation_button = driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input')
designation_button.send_keys('Data Scientist')
```

```python
#click the submit button to show the list of jobs
submit = driver.find_element(By.XPATH,'//div[@class="qsbSubmit"]')
submit.click()
time.sleep(5)
```

```python
#Apply location filter by checking the respective box
location_filter = driver.find_element(By.XPATH,'//div[@data-filter-id="citiesGid"]/div[2]/label/i')
location_filter.click()
```

```python
#Apply Salary filter by checking the respective box
salary_filter = driver.find_element(By.XPATH,'//div[@data-filter-id="salaryRange"]/div[2]/label/i')
salary_filter.click()
```

```python
#create the empty list for the job title, location, company, experience
title=[]
location=[]
company=[]
experience=[]

#Get the company name,title,location,experience for first 10 job search
com=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[2]/span/a[1]')
for i in com:
    company.append(i.text.strip())
    company=company[0:10]
header=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]')
for i in header:
    jtitle= i.find_element(By.CLASS_NAME,'title ')
    title.append(jtitle.text.strip())
    title=title[0:10]
    loc=i.find_element(By.CLASS_NAME,'locWdth')
    location.append(loc.text.strip())
    location=location[0:10]
    exp=i.find_element(By.CLASS_NAME,'expwdth')
    experience.append(exp.text.strip())
    experience = experience[0:10]

#create the dataframe with the job-title,job-location,company_name,experience
df=pd.DataFrame({"Job-title":title,"job-location":location,"company-name":company,"experience-required":experience})
df
```

Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: 
https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART

As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews.

Note: All the stepsrequired during scraping should be done through code only and not manually.

```python
#Instantiate the webdriver for Chrome
driver=webdriver.Chrome()
```

```python
#Get the URL using get() method
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
```

```python
#Maximize the window size of chrome
driver.maximize_window()
```

```python
#Create Empty list for rating,review,fullreview
rating=[]
review=[]
fullreview=[]

#Get the rating,review,fullreview for first 100 iphone search
for i in range(1,11):
    ratings = driver.find_elements(By.XPATH,'//div[@class="EKFha-"]')
    for i in ratings:
        rating.append(i.text.strip().split("\n")[0])
        rating=rating[0:100]
        review.append(i.text.strip().split("\n")[1])
        review=review[0:100]
        fullreview.append(i.text.strip().split("\n")[2])
        fullreview=fullreview[0:100]
    next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(5)

#create the dataframe with rating,review,fullreview for 100 iphone search
df1=pd.DataFrame({"ratings":rating,"reviews":review,"fullreview":fullreview})
df1
```

2. Click on TopQuote
3. Than scrap 

a)Quote 

b) Author 

c) Type Of Quotes

```python
#Instantiate the webdriver for Chrome
driver=webdriver.Chrome()
```

```python
#Get the URL using get() method
driver.get("https://www.azquotes.com/")
```

```python
#Maximize the window size of chrome
driver.maximize_window()
```

```python
#click the top quotes icon
top_quotes=driver.find_element(By.XPATH,'//div[@class="mainmenu"]/ul/li[5]/a')
top_quotes.click()
```

```python
#Create Empty list for quote,author,typeofquotes
quote=[]
author=[]
typeofquotes=[]

#Get the quote,author and typeofquotes for first 1000 quotes
for i in range(0,10):
    top=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in top:
        quote.append(i.text.strip())
        quote=quote[0:1000]
    authors=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in authors:
        author.append(i.text.strip())
        author=author[0:1000]
    typeofquote=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in typeofquote:
        typeofquotes.append(i.text.strip())
        typeofquotes=typeofquotes[0:1000]       
    next_btn=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    next_btn.click()

#Create the dataframe with Quotes,Author and Type Of Quotes for first 1000 quotes searcch
df=pd.DataFrame({"Quotes":quote,"Author":author,"Type Of Quotes":typeofquotes})
df
```

Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1

scrap the mentioned data and make the DataFrame

```python
#Instantiate the webdriver for Chrome
driver=webdriver.Chrome()
```

```python
#Get the URL using get() method
driver.get("https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1")
```

```python
#Maximize the window size of chrome
driver.maximize_window()
```

```python
#click on the Accept button to accept terms of the page
accept=driver.find_element(By.XPATH,'//div[@class="jsx-ceb768ea3afd33f5 closegdper btn"]')
accept.click()
```

```python
#Create Empty list for name,borndead,term,remarks
name=[]
born_dead=[]
term=[]
remarks=[]

#Get the name,borndead, term of office and remarks for former prime minister of India
names=driver.find_elements(By.XPATH,'//div[@class="TableData"]//tr/td[2]')
for i in names:
    name.append(i.text.strip())
    name=name[0:19]
borndead=driver.find_elements(By.XPATH,'//div[@class="TableData"]//tr/td[3]')
for j in borndead:
    born_dead.append(j.text.strip())
    born_dead=born_dead[0:19]
terms=driver.find_elements(By.XPATH,'//div[@class="TableData"]//tr/td[4]')
for k in terms:
    term.append(k.text.strip().split("\n")[0].split(",")[0])
    term=term[0:19]
remark=driver.find_elements(By.XPATH,'//div[@class="TableData"]//tr/td[5]')
for l in remark:
    remarks.append(l.text.strip())
    remarks=remarks[0:19]

#Create the dataframe with Name,borndead,termofoffice,remarks for the former prime minister of India
df=pd.DataFrame({"Name":name,"Born-Dead":born_dead,"Term of office":term,"Remarks":remarks})
df
```

Q8: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/

    This task will be done in following steps:    
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ‚Äô50 most expensive cars‚Äô
3. Then click on 50 most expensive carsin the world..
4. Then scrap thementioned data and make the dataframe

```python
#Instantiate the webdriver for Chrome
driver=webdriver.Chrome()
```

```python
#Get the URL using get() method
driver.get("https://www.motor1.com/")
```

```python
#Maximize the window size of chrome
driver.maximize_window()
```

```python
#Click the search icon
search=driver.find_element(By.XPATH,'//form[@class="m1-search-form"]')
search.click()
```

```python
#Search for 50 most expensive cars in the search panel
search_panel=driver.find_element(By.XPATH,'//input[@class="m1-search-panel-input m1-search-form-text"]')
search_panel.send_keys("50 most expensive cars")
search_panel.send_keys(Keys.ENTER)
```

```python
#click the '50 most expensive cars in the world' icon
click=driver.find_element(By.XPATH,'//div[@class="item wcom "]/div/div/h3/a')
click.click()
```

```python
#Create Empty list for carname,price
carname=[]
price=[]

#Get the car name and price for the 50 expensive cars in the world
cars=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in cars:
    carname.append(i.text.strip())
    carname=carname[0:50]
prices=driver.find_elements(By.XPATH,'//div[@class="postBody description e-content"]/p/strong')
for j in prices:
    price.append(j.text.strip().replace('Price: ',''))

#Create the dataframe with car name,price for the 50 expensive cars in the world
df=pd.DataFrame({"Car name":carname,"Price":price})
df
```

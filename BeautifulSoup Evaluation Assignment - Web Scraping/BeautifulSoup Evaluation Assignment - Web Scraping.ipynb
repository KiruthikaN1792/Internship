1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
def top_100_imdb_movies(url):
    response = requests.get(url)
    sp = BeautifulSoup(response.content)
    name=[]
    ratings=[]
    yearofrelease=[]
    for i in sp.find_all('h3',{"class":"lister-item-header"}):
        name.append(i.text.replace("\n","").strip().split('.')[1].split('(')[0])
        name=name[0:100]
    for j in sp.find_all('div',{"class":"ipl-rating-star small"}):
        ratings.append(j.text.replace("\n",""))
        ratings=ratings[0:100]
    for k in sp.find_all('span',{"class":"lister-item-year text-muted unbold"}):
        yearofrelease.append(k.text.replace("\n","").strip('()'))
        yearofrelease=yearofrelease[0:100]
    
    df= pd.DataFrame({"name":name,"rating":ratings,"year ofrelease":yearofrelease})
    return df
```

```python
top_100_imdb_movies("https://www.imdb.com/list/ls056092300/")
```

5) Write a python program to scrape house details from mentioned URL. It should include house title, location, area, EMI and price from https://www.nobroker.in/ .Enter three localities which are Indira Nagar, Jayanagar,
Rajaji Nagar.

```python
def get_nobroker(url):
    res=requests.get(url)
    bp=BeautifulSoup(res.content,'html.parser')
    house=[]
    location=[]
    area=[]
    emi=[]
    prices=[]
    for i in bp.find_all('a',{"class":"overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"}):
        house.append(i.text.strip().split(',')[0])
    for j in bp.find_all('div',{"class":"mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95"}):
        location.append(j.text.strip())
    for m in bp.find_all('div',{"id":"unitCode"}):
        area.append(m.text.strip())
    for k in bp.find_all('div',{"id":"roomType"}):
        emi.append(k.text.strip())
    top=bp.find_all('div',{"class":"nb__7nqQI"})
    for i in top:
        prc=bp.find('div',{"class":"font-semi-bold heading-6"})
        prices.append(prc.text.strip().split(',')[0])
    df=pd.DataFrame({"house_title":house,"location":location,"Area":area,"EMI":emi,"Price":prices})
    return df
```

6) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .

```python
def get_10_product_bewakoof(url):
    response = requests.get(url)
    bs=BeautifulSoup(response.content)
    product=[]
    price=[]
    img=[]
    for i in bs.find_all('div',{"class":"productNaming bkf-ellipsis"}):
        product.append(i.text.strip())
        product=product[0:10]
    for j in bs.find_all('div',{"class":"discountedPriceText"}):
        price.append(j.text.strip())
        price=price[0:10]
    for k in bs.find_all('img',{"class":"productImgTag"}):
        img.append(k["src"].strip())
    df1=pd.DataFrame({"Product name":product,"Price":price,"Image URL":img})
    return df1
```

```python
get_10_product_bewakoof("https://www.bewakoof.com/bestseller?sort=popular")
```

7) Please visit https://www.cnbc.com/world/?region=world and scrapa)
headings
b) date
c) News link

```python
def get_cnbc_news(url):
    resp=requests.get(url)
    soup = BeautifulSoup(resp.content)
    headings=[]
    datess=[]
    newslinks=[]
    top=soup.find_all('div',{"class":"RiverPlusCard-cardLeft"})
    for i in top:
        heading=i.find('a').text.strip()
        date=i.find('div',{"class":"RiverByline-datePublished"})
        datetime=date.text.strip() if date else " "
        news = i.find('a')["href"]
        headings.append(heading)
        datess.append(datetime)
        newslinks.append(news)
    df2=pd.DataFrame({"headings":headings,"dates":datess,"Newslink":newslinks})
    return df2
```

```python
get_cnbc_news("https://www.cnbc.com/world/?region=world")
```

8) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/
and scrap
a)Paper title
b) date
c) Author

```python
def get_kaei_ai(url):
    res=requests.get(url)
    bsoup = BeautifulSoup(res.content)
    papers=[]
    datess=[]
    authors=[]
    for i in bsoup.find_all('h2',{"class":"h5 article-title"}):
        papers.append(i.text.strip())
    for j in bsoup.find_all('p',{"class":"article-date"}):
        datess.append(j.text.strip().split(' ')[-1])
    for k in bsoup.find_all('p',{"class":"article-authors"}):
        authors.append(k.text.strip())
    data={"Paper title":papers,"date":datess,"Author":authors}
    df3=pd.DataFrame(data)
    return df3
```

```python
get_kaei_ai("https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/")
```

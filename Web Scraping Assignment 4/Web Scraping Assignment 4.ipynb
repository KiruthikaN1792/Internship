1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 

A)Rank 

B) Name 

C) Artist 

D) Upload date 

E) Views

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(3)
```

```python
driver.maximize_window()
```

```python
rank=[]
name=[]
artist=[]
uploaddate=[]
views=[]

try:
    header=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for i in header:
        name.append(i.text.strip().split("[")[0])
except NoSuchElementException:
    name.append('-')
try:
    artists=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for j in artists:
        artist.append(j.text.strip())
except NoSuchElementException:
    artist.append('-')
try:
    view=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for k in view:
        views.append(k.text.strip())
except NoSuchElementException:
    views.append('-')
try:
    date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for l in date:
        uploaddate.append(l.text.strip())
except NoSuchElementException:
    uploaddate.append("-")

df=pd.DataFrame({"Name":name,"Artist":artist,"Upload Date":uploaddate,"Views":views})
df
```

5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the following details: 

A) Song name 

B) Artist name 

C) Last week rank 

D) Peak rank 

E) Weeks on board 

Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
driver=webdriver.Chrome()
```

```python
driver.get("https:/www.billboard.com/")
```

```python
driver.maximize_window()
```

```python
chart=[]
headers=[]
songs=[]
music=[]
artists=[]
lastweek=[]
peak=[]
weeksonboard=[]

charts=driver.find_element(By.XPATH,'//h3[@id="mega-menu-item-charts"]/a')
chart.append(charts.get_attribute('href'))
for i in chart:
    driver.get(i)
    song=driver.find_element(By.XPATH,'//div[@class="lrv-u-flex lrv-u-justify-content-center"]/a')
    songs.append(song.get_attribute('href'))
    for j in songs:
        driver.get(j)
        try:
            header=driver.find_elements(By.XPATH,'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div/ul/li[4]/ul/li[1]/h3')
            for k in header:
                music.append(k.text)
        except NoSuchElementException:
            music.append('-')
        try:
            artist=driver.find_elements(By.XPATH,'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div/ul/li[4]/ul/li[1]/span')
            for l in artist:
                artists.append(l.text)
        except NoSuchElementException:
            artists.append('-')
        try:
            lastweeks=driver.find_elements(By.XPATH,'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div/ul/li[4]/ul/li[4]/span')
            for m in lastweeks:
                lastweek.append(m.text.strip())
        except NoSuchElementException:
            lastweek.append('-')
        try:
            peaks=driver.find_elements(By.XPATH,'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div/ul/li[4]/ul/li[5]/span')
            for o in peaks:
                if o.text is not None:
                    peak.append(o.text)
        except NoSuchElementException:
            peak.append('-')
        try:
            weeks=driver.find_elements(By.XPATH,'//*[@id="post-1479786"]/div[3]/div/div/div/div[2]/div/ul//li[6]/span')
            for n in weeks:
                weeksonboard.append(n.text)
        except NoSuchElementException:
            weeksonboard.append('-')
            
df=pd.DataFrame({"Song name":music,"Artist name":artists,"Last week rank":lastweek,"Peak rank":peak,"Weeks on board":weeksonboard})
df
```

```python
dataset=driver.find_element(By.XPATH,'//div[@class="flex flex-wrap justify-center gap-5"]/a[1]')
dataset.click()
time.sleep(5)
```

```python
dataset=[]
datatype=[]
task=[]
attributetype=[]
instances=[]
attributes=[]
year=[]

for i in range(0,60):
    collapse=driver.find_element(By.XPATH,'//label[@class="btn-primary btn-sm btn flex gap-2 rounded-full"]/div[2]/span[1]')
    collapse.click()
    time.sleep(8)
    expand=driver.find_element(By.XPATH,'//label[@class="btn-primary btn-sm btn flex gap-2 rounded-full"]/div[2]/span[2]')
    expand.click()
    time.sleep(8)
    try:
        datasets=driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/h2/a')
        for i in datasets:
            dataset.append(i.text)
    except NoSuchElementException:
        dataset.append('-')
    try:
        datatypes=driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[2]/span')
        for i in datatypes:
            datatype.append(i.text)
    except NoSuchElementException:
        datatype.append('-')
    try:
        tasks=driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[1]/span')
        for i in tasks:
            task.append(i.text)
    except NoSuchElementException:
        task.append('-')
    try:
        instance=driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[3]/span')
        for i in instance:
            instances.append(i.text.split(' ')[0])
    except NoSuchElementException:
        instances.append('-')
    try:
        attribute=driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[4]/span')
        for i in attribute:
            attributes.append(i.text.split(' ')[0])
    except NoSuchElementException:
        attributes.append('-')
    try:
        types=driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]//tbody/tr/td[2]')
        for i in types:
            attributetype.append(i.text)
    except NoSuchElementException:
        attributetype.append('-')
    try:
        years=driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]//tbody/tr/td[3]')
        for i in years:
            year.append(i.text)
    except NoSuchElementException:
        year.append('-')
    next_button=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
    next_button.click()
df=pd.DataFrame({"Dataset name":dataset,"Data type":datatype,"Task":task,"Attribute type":attributetype,"No of Instances":instances,"No of Attributes":attributes,"Year":year})
df
```
